```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1) #10% test & 90% train mya jaygi
X_train #independent data
```

```python
Y_train  #dependent data
```

```python
Y_test
```

```python
X_test
```

```python
#Scale the value
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() #object = class
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf =  LogisticRegression() #Create Model
```

```python
#model training
clf.fit(X_train,Y_train)
```

```python
#Evaluate the model
Y_pred = clf.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# Plotting decision regions
plot_decision_regions(X_train,Y_train.values, clf=clf, legend=2)  #classifier -> clf
```
